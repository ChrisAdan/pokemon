```python
def fetch_all_moves():
    '''Fetch all available move names from the PokeAPI REST.'''

    url = 'https://pokeapi.co/api/v2/move?limit=1000'
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        return [move['name'] for move in data['results']]
    else:
        raise Exception('Failed to fetch move list')
```

```python
with open(f'../queries/extract_types.gql', 'r') as file:
    query = ''.join(line for line in file if not line.strip().startswith('#'))

def get_type_matchup(primary_type, secondary_type):
    variables = {
        'primaryType': primary_type.lower(),
        'secondaryType': secondary_type.lower()
    }
    response = requests.post(url, json={'query': query, 'variables': variables}, headers=headers)
    return response.json()

types = [
    'Normal', 'Fire', 'Water', 'Electric', 'Grass', 'Ice', 'Fighting', 'Poison', 
    'Ground', 'Flying', 'Psychic', 'Bug', 'Rock', 'Ghost', 'Dragon', 'Dark', 
    'Steel', 'Fairy'
]

for primary_type in types:
    for secondary_type in types:
        if primary_type == secondary_type:
            continue
        data = get_type_matchup(primary_type, secondary_type)
        if 'data' in data and 'getTypeMatchup' in data['data']:
            matchup = data['data']['getTypeMatchup']

            attacking = matchup['attacking']
            defending = matchup['defending']
```
